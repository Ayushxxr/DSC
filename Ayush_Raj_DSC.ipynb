 
# Inference

It is used to import seaborn library by which we can make many statistical graphics like Histogram,Boxplot etc.

```python

plt.figure(figsize=(10,6))
sns.histplot(data['Exam_Score'], kde=True, bins=20)

plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')

plt.show()
```

# Inference

The above Histogram shows Frequency vs Exam score from which we can ellaborate that highest bar indicates the mode where highest frequency of exam score is located also from the above plot we can depict that exam scores are spread out from around 50 to a little over 100 despite of the fact that most scores are concentrated around 60-70 range. Also there are very few no of students who scored higher than 80 which means that very high score are rare in the dataset.  Also the histogram appears to be left skewed meaning that most of the exam scores are clustered towards the higher end

```python
# Boxplot to visualize outliers in the Exam Score
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Exam_Score'])

# Add labels and title
plt.title('Box Plot of Exam Scores')
plt.xlabel('Exam Score')

# Show the plot
plt.show()
```

# Inference

This box plot give us an idea about the median which is central line in the box,
this box tell us about the IQR which is from first quartile to third quartile.
Also the two lines(Whiskers) extending right and left to the box tells about the range in which most of the non outlier data points lie and finally the dots outside the two lines tell us about the outliers



```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for Exam_Score
Q1 = data['Exam_Score'].quantile(0.25)
Q3 = data['Exam_Score'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = data[(data['Exam_Score'] < lower_bound) | (data['Exam_Score'] > upper_bound)]
print("Outliers in Exam Score:")
print(outliers)
```

# Inference

Outliers are data points that are very differet from other as they are very higher or lower than most of the other values in dataset

```python
# Calculate Mean
mean_exam_score = data['Exam_Score'].mean()
print(f"Mean of Exam Score: {mean_exam_score:.2f}")

# Calculate Median
median_exam_score = data['Exam_Score'].median()
print(f"Median of Exam Score: {median_exam_score:.2f}")

# Calculate Standard Deviation
std_exam_score = data['Exam_Score'].std()
print(f"Standard Deviation of Exam Score: {std_exam_score:.2f}")
```

# Inference

Mean provides central value for data and helpful when we compare different datset
Median gives middle value of a dataset when it is in ascending or descending order such that 50% of data points are below and above
Standard deviation tell us about the variation in the datset


```python
# Calculate mean and median of Exam_Score
mean_score = data['Exam_Score'].mean()
median_score = data['Exam_Score'].median()

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Exam_Score'], kde=True, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean_score, color='red', linestyle='--', linewidth=1.5, label=f'Mean: {mean_score:.2f}')
plt.axvline(median_score, color='green', linestyle='-', linewidth=1.5, label=f'Median: {median_score:.2f}')
plt.title('Exam Score Distribution with Mean and Median')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Inference

This histogram shows the distribution of exam scores along with the mean and median
The red dashed line represent the mean score which is 67.24 and green line represent median score which is 67, The highest bar or mode is around the 67 mark, meaning most students scored close to this value. Most exam scores are between 60 and 75 and very few score beyond 75.This tells us that extreme high or low scores are rare in this dataset.The few points beyond 80 can be considered as outliers
